```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import vonmises # von Mises distribution
from scipy.special import iv # Bessel Function
```

```python

# First von Mises density's parameters
m1 = 5  # concentration
theta01 = 0.25 * np.pi  # circular mean

# Second von Mises density's parameters
m2 = 1  # concentration
theta02 = 0.75 * np.pi  # circular mean

# The number of sample size
number_of_samples = 1000

# Edit x-limits of Cartesian plot
left_lim = 0
right_lim = 2 * np.pi
```

```python
x = np.linspace(left_lim, right_lim, 500)
# First von Mises density
vonmises_pdf1 = vonmises.pdf(theta01, m1, x)
# Second von Mises density
vonmises_pdf2 = vonmises.pdf(theta02, m2, x)
```

```python
fig = plt.figure(figsize=(12, 6))

# Left : Cartesian plot
left = plt.subplot(121)
left.plot(x, vonmises_pdf1, label= r'$m=5, \theta_0=\pi/4$')
left.plot(x, vonmises_pdf2, label= r'$m=1, \theta_0=3\pi/4$')
left.set_title("Cartesian plot")
left.set_xlim(left_lim, right_lim)
left.grid(True)
left.legend()

# Right : Polar plot
right = plt.subplot(122, projection='polar')
right.plot(x, vonmises_pdf1, label= r'$m=5, \theta_0=\pi/4$')
right.plot(x, vonmises_pdf2, label= r'$m=1, \theta_0=3\pi/4$')
right.set_title("Polar plot")
right.legend()
```

```python
x = np.linspace(0, 10, 100)
Am = iv(1, x) / iv(0, x) # Function A(m)
```

```python
# Left : Modified Bessel functions of the first kind I0(m)
fig = plt.figure(figsize=(12, 6))
left = fig.add_subplot(121)
left.set_title("Modified Bessel functions of the first kind", size=15)
left.grid()
left.set_xlim(0, 10)
left.set_ylim(0, 3000)
left.set_xlabel(r'$m$', size=15, labelpad=10)
left.set_ylabel(r'$I_0(m)$', size=15, labelpad=8)
left.plot(x, iv(0, x))

# Right : Function A(m)
right = fig.add_subplot(122)
right.set_title(r'Function A(m)=$I_1(m) / I_0(m)$', size=15)
right.grid()
right.set_xlim(0, 10)
right.set_ylim(0, 1)
right.set_xlabel(r'$m$', size=15, labelpad=10)
right.set_ylabel(r'$A(m)$', size=15, labelpad=8)
right.plot(x, Am)
```
