```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Edit and add a sample_size
sample_size = [1, 2, 10]

# Edit the number of simulations
sim = 10000
```

```python
# Function generates uniform random numbers
def generate_uniform_numbers(size, low, high):
    return np.random.uniform(low, high, size)

# Function plots histogram of means of uniform random numbers
def plot_histogram(means, title):
    plt.hist(means, bins='auto', alpha=0.7, rwidth=0.85)
    plt.xlabel('Mean')
    plt.ylabel('Frequency')
    plt.title(title)
    plt.show()

for i in range(len(sample_size)):
    random_numbers = []
    for j in range(sim):
        random_set = np.mean(generate_uniform_numbers(sample_size[i], 0, 1))
        random_numbers.append(random_set)
    plot_histogram(random_numbers, f"Histogram of Means (N = {sample_size[i]})")
```
