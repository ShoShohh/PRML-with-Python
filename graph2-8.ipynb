```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# The mean vector and covariance matrix
mean = np.array([0, 0])  # Mean vector
covariance = [
    np.array([[5, 2], [2, 1]]),  # Covariance matrix 1 (regular)
    np.array([[5, 0], [0, 1]]),  # Covariance matrix 2 (diagonal matrix)
    np.array([[1, 0], [0, 1]])  # Covariance matrix 3 (identity matrix)
]
```

```python
# Grid points for x and y axes
x_1 = np.linspace(-5, 5, 100)
x_2 = np.linspace(-5, 5, 100)
X_1, X_2 = np.meshgrid(x_1, x_2)

# The probability density function (PDF)
pos = np.dstack((X_1, X_2))
for i in range(len(covariance)):
    Z = np.exp(-0.5 * np.einsum('ijk,kl,ijl->ij', pos - mean, np.linalg.inv(covariance[i]), pos - mean)) / (2 * np.pi * np.sqrt(np.linalg.det(covariance[i])))
    # Plot the contour lines
    plt.contour(X_1, X_2, Z)
    plt.axis('square')
    plt.xlabel('$X_1$')
    plt.ylabel('$X_2$')
    plt.title('Contour Plot of a 2D Gaussian Distribution')
    plt.text(3.5, 4, s = f"{covariance[i]}")
    plt.show()
```
